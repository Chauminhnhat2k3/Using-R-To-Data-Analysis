```r
library(tidyverse)
```

```r
data <- read.csv("gapminder-FiveYearData.csv")
print(data)
```

```r
library(dplyr)
```

```r
head(data)
```

```r
#1.	Filter the data for the Americas in 2007, deselect all other variables
data1 = filter(data,continent == "Americas" & year == 2007)
print(data1)
```

```r
#2.	Create the variable gdp, defined as the product of population size and gdp per person.
population_size <- data$pop
gdp_per_person <- data$gdpPercap

gdp <- population_size * gdp_per_person

print(gdp)
```

```r
#3.	Identify the observation with lowest gdp per person.
lowest_gdp_person <- which.min(data$gdpPercap)
lowest_gdp_person
```

```r
#4.	Identify all observations with above average life expectancy, stratified for each continent.
# Calculate global average life expectancy
global_avg_le <- mean(data$lifeExp)

# Filter for observations with above average life expectancy, stratified by continent
above_avg_le_by_continent <- data %>%
  group_by(continent) %>%
  filter(lifeExp > global_avg_le) %>%
  ungroup()

# View the resulting data frame
above_avg_le_by_continent
```

```r
#5.	Count the observations identified in the last step.
nrow(above_avg_le_by_continent)
```

```r
#6.	Compute the mean life expectancy (the grand mean; ie., across all observations).
grand_mean_le <- mean(data$lifeExp)

grand_mean_le
```

```r
#7.	Compute the mean life expectancy for each year.
mean_le_by_year <- data %>%
  group_by(year) %>%
  summarize(mean_life_exp = mean(lifeExp))

mean_le_by_year
```
